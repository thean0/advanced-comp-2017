```python
%config InlineBackend.figure_format='retina'
%matplotlib inline

import numpy as np
np.random.seed(123)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8, 8)
plt.rcParams["font.size"] = 14
from sklearn.utils import check_random_state
```

Creating my random dataset:

```python
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier

labels = ["b", "r"]
X, y = make_blobs(n_samples=400, centers=23, random_state=42)
y = np.take(labels, (y < 10))
```

Plotting my dataset

```python
plt.scatter(X[:,0], X[:,1], c=y, lw=0, s=40)
```

We do not know yet, how the ratio of the training dataset to the validation data influences the model, so we decided  to first try it with a 50-50% devision.

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.5)
```

First we want to determine the optimal splitting ratio, and then the optimal n_neigbour number. So first we fix the k neighbour number of the model as k = 10.

```python
clf = KNeighborsClassifier(n_neighbors = 10)
clf.fit(X_train, y_train)
```

Evaluating the performance of the classifier on the testing and then on the validation dataset.

```python
train_scores = clf.score(X_train, y_train)
test_scores = clf.score(X_valid, y_valid)

print("Train score: %.4f" % train_scores)
print("Test score: %.4f" % test_scores)
```

Repeating this for various splitting ratio, and also averaging for different data realization to get a smoother result

```python
acc_T = []
acc_V = []
splits = np.arange(0.1, 1.0, 0.1)

for n in range(50):
    X, y = make_blobs(n_samples=400, centers=23, random_state=42+n)
    y = np.take(labels, (y < 10))
    train_scores = []
    test_scores = []
    
    for sp in splits:
        X_train,X_test, y_train,y_test = train_test_split(X, y, train_size=sp)
        clf = KNeighborsClassifier(n_neighbors=10)
        clf.fit(X_train, y_train)
        train_scores.append(clf.score(X_train, y_train))
        test_scores.append(clf.score(X_test, y_test))
        
    acc_T.append(test_scores)
    acc_V.append(train_scores)
    
    plt.plot(splits, train_scores, c='b', alpha=0.1)
    plt.plot(splits, test_scores, c='r', alpha=0.1)
    
plt.plot(splits, np.array(acc_T).mean(axis=0), label='Test', c='r', lw=4)
plt.plot(splits, np.array(acc_V).mean(axis=0), label='Train', c='b', lw=4)
plt.xlabel('Splitting ratio')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```

1) The training results are always higher, which is natural, since we are training the model on that data

2) When we choose 90% of our data as training data, we overfit the result. The scores on the training set are going to improve, but this will not improve the score on the validation data, because our model is specialized for the training set.

3) The best splitting ratio would be to choose the point where the plateau starts: around 50-60%

```python
acc_T = []
acc_V = []
k_Nb = np.arange(1, 20, 1)

for n in range(50):
    X, y = make_blobs(n_samples=400, centers=23, random_state=42+n)
    y = np.take(labels, (y < 10))
    X_train,X_test, y_train,y_test = train_test_split(X, y, train_size=0.6)
    train_scores = []
    test_scores = []
    
    for k in k_Nb:
        clf = KNeighborsClassifier(n_neighbors=k)
        clf.fit(X_train, y_train)
        train_scores.append(clf.score(X_train, y_train))
        test_scores.append(clf.score(X_test, y_test))
        
    acc_T.append(test_scores)
    acc_V.append(train_scores)
    
    plt.plot(k_Nb, train_scores, c='b', alpha=0.1)
    plt.plot(k_Nb, test_scores, c='r', alpha=0.1)
    
plt.plot(k_Nb, np.array(acc_T).mean(axis=0), label='Test', c='r', lw=4)
plt.plot(k_Nb, np.array(acc_V).mean(axis=0), label='Train', c='b', lw=4)
plt.xlabel('k - Inverse model complexity')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```

1) The training results are always higher, which is natural, since we are training the model on that data

2) In case of low k, the model is more complex, it memorizes the inputs from the training dataset. So the performance on that is rapidly increasing, but this means that on unseen datapoints the performance is going to decrease.

3) The best fit is to choose a point after the two graphs take different routes: increasing and decreasing performance. This is around 7 neighbours. Also there is a plateau phase after that. Which means, that decreasing the complexity of the model will not reduce the performace. It's worth to work with a more simple model, so I would choose a value around the end of the plateau: 11.
